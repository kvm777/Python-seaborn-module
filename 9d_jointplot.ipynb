```python
d2 = pd.read_csv("C:\\Users\\korad\\Desktop\\seaborn\\188.iris.csv")
d2.head()
```

```python
sns.jointplot(data=d2,x='SepalLength',y='PetalLength',marginal_kws=dict(color='green',kde=True))
```

```python
sns.jointplot(data=d2,x='SepalLength',y='PetalLength',kind='hist')
```

```python
sns.jointplot(data=d2,x='SepalLength',y='PetalLength',kind='kde')
```

```python
sns.jointplot(data=d2,x='SepalLength',y='PetalLength',kind='hex')
```

```python
sns.jointplot(data=d2,x='SepalLength',y='PetalLength',kind='reg')
```

```python
sns.jointplot(data=d2,x='SepalLength',y='PetalLength',kind='resid')
```

```python
l=sns.jointplot(data=d2,x='SepalLength',y='PetalLength',joint_kws=dict(color='red'))
l.plot_joint(sns.kdeplot)
l.plot_joint(sns.rugplot,height=0.1)
```
