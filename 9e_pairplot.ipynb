```python
#pairplot

import numpy as np
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
```

```python
d1 = pd.read_csv("C:\\Users\\korad\\Desktop\\seaborn\\188.iris.csv")
d1.head()
```

```python
sns.pairplot(data=d1)
```

```python
sns.pairplot(data=d1,kind='reg')
```

```python
sns.pairplot(data=d1,corner=True)
```

```python
sns.pairplot(data=d1,diag_kws=dict(color='violet',kde=True),plot_kws=dict(color='red'))
```

```python
l = sns.pairplot(data=d1,plot_kws=dict(color='red'))
l.map_upper(sns.kdeplot)
```

```python
l = sns.pairplot(data=d1,plot_kws=dict(color='red'))
l.map_lower(sns.kdeplot)
```
