```python
#displot..... Distribution plot

import numpy as np
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
```

```python
d1=pd.read_csv("C:\\Users\\korad\\Desktop\\seaborn\\187.tips.csv")
d1.head()
```

```python
#by default histogram is created.

sns.displot(data=d1,x='total_bill',kde=True,rug=True,rug_kws={'height':0.05})
```

```python
sns.displot(data=d1,x='total_bill',kde=True,rug=True,rug_kws={'height':0.05},hue='day',multiple='stack',col='sex',row='smoker')
```

```python
sns.displot(data=d1,x='total_bill',hue='day',kind='ecdf',col='sex',row='smoker')
```
